```python
from gtts import gTTS
import os

def text_to_speech(text, language='hi', output_file='output.mp3'):
    tts = gTTS(text=text, lang=language, slow=False)
    tts.save(output_file)
    os.system(f'start {output_file}')  # Plays the generated speech using the default media player

# Example usage:
text = "5 महत्वपूर्ण जीवन के पाठ जो भगवद गीता से हर एक इंसान को सीखने चाहिए।  1. आपको अपने काम का आनंद लेना चाहिए  2. आपको अपनी भावनाओं को Manage करना होगा  3. किसी और के जीवन की नकल करने की कोशिश न करें   4.अपने लक्ष्यों को हमेशा याद रखें    5. ज्यादा सोचो मत, Action लो"
text_to_speech(text)
```

```python
from gtts import gTTS
import os

def download_text_to_speech(text, language='en', output_file='output(1).mp3'):
    tts = gTTS(text=text, lang=language, slow=False)
    tts.save(output_file)
    print(f"Text-to-speech generated and saved as {output_file}")

# Example usage:
text = "Hello, and welcome back to our channel! Today, we're excited to share ten valuable tips that will help boost your productivity and make the most of your time."
download_text_to_speech(text, language='en', output_file='speech.mp3')
```

```python
from moviepy.editor import TextClip, CompositeVideoClip, concatenate_videoclips

def text_to_video(text, output_file='output.mp4', duration_per_slide=5, bg_color='white', txt_color='black', fontsize=40):
    # Create a TextClip for each slide of the text
    text_clips = [TextClip(text, fontsize=fontsize, color=txt_color, bg_color=bg_color, size=(1920, 1080)).set_duration(duration_per_slide)
                  for text in text.split('\n\n')]

    # Concatenate all the text clips to create the video
    final_clip = concatenate_videoclips(text_clips, method='compose')

    # Export the video
    final_clip.write_videofile(output_file, fps=30, codec='libx264')

# Example usage:
text = """
Slide 1: Welcome to our video!
This is a sample text-to-video conversion using Python and MoviePy.

Slide 2: MoviePy is awesome!
MoviePy is a fantastic library for video editing in Python.

Slide 3: End of the video
Thank you for watching! Don't forget to like and subscribe!
"""

text_to_video(text, output_file='text_video.mp4')
```
